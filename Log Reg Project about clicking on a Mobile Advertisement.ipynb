```python
#Lets do the Regression on our model
#Let's do the train test split & Train our model
#Split the data into training set and testing set using train_test_split
```

```python
ad_data.columns
```

```python
X= ad_data[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage','Male']]
```

```python
Y= ad_data[['Clicked on Ad']]
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3, random_state=100)
```

```python
#Train & Fit a Logistic model on the Training Set
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrmodel= LogisticRegression()
```

```python
lrmodel.fit(X_train,Y_train)
```

```python
#Prediction & Evaluations
```

```python
Prediction= lrmodel.predict(X_test)
```

```python
#Lets Make A Classification Report for the model
```

```python
from sklearn import metrics
```

```python
print (metrics.classification_report(Y_test,Prediction))
```
